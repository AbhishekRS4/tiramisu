```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrics_103 = np.load('model_103_100/metrics.npy').item()
metrics_67 = np.load('model_67_100/metrics.npy').item()
metrics_56 = np.load('model_56_100/metrics.npy').item()
```

```python
x_values = np.arange(1, len(metrics_103['train_loss']) + 1)

plt.figure(figsize = (12, 8))
plt.title('loss vs epoch')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.plot(x_values, metrics_103['train_loss'], 'b--', label = 'tiramisu-103 training loss')
plt.plot(x_values, metrics_103['valid_loss'], 'r--', label = 'tiramisu-103 validation loss')

plt.plot(x_values, metrics_67['train_loss'], 'g--', label = 'tiramisu-67 training loss')
plt.plot(x_values, metrics_67['valid_loss'], 'c--', label = 'tiramisu-67 validation loss')

plt.plot(x_values, metrics_56['train_loss'], 'm--', label = 'tiramisu-56 training loss')
plt.plot(x_values, metrics_56['valid_loss'], 'k--', label = 'tiramisu-56 validation loss')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.title('accuracy vs epoch')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.plot(x_values, metrics_103['valid_acc'], 'b--', label = 'tiramisu-103 validation accuracy')
plt.plot(x_values, metrics_67['valid_acc'], 'g--', label = 'tiramisu-67 validation accuracy')
plt.plot(x_values, metrics_56['valid_acc'], 'r--', label = 'tiramisu-56 validation accuracy')

plt.legend(loc = 4)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.title('mean IoU vs epoch')
plt.xlabel('epoch')
plt.ylabel('mean IoU')

plt.plot(x_values, metrics_103['valid_iou'], 'b--', label = 'tiramisu-103 validation mean IoU')
plt.plot(x_values, metrics_67['valid_iou'], 'g--', label = 'tiramisu-67 validation mean IoU')
plt.plot(x_values, metrics_56['valid_iou'], 'r--', label = 'tiramisu-56 validation mean IoU')

plt.legend(loc = 4)
plt.show()
```
